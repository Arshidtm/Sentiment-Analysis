```python
stemmer=SnowballStemmer('english')
```

```python
def clean(text): 
    text=text.lower()
    # removing charachter other than alphabets
    text=re.sub('[^a-zA-Z]',' ',text)
    token=word_tokenize(text)
    text=[stemmer.stem(word) for word in token ]
    return ' '.join(text)
    
    
```

```python
df['Cleaned Review']=df['review'].apply(clean)
```

```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
```

```python
simple_preprocess(df['Cleaned Review'].iloc[0])
```

```python
words=[] 
for sent in df['Cleaned Review']: 
    words.append(simple_preprocess(sent))
    
```

```python
model4=Word2Vec(window=5,min_count=3)
model4.build_vocab(words)
model4.train(words,total_examples=model4.corpus_count,epochs=30)
```

```python
def avg_word2veec(doc): 
    tok=[word for word in doc if word in model4.wv.index_to_key]
    if not tok: 
        return np.zeros(model4.vector_size)
    return np.mean(model4.wv[tok],axis=0)
```

```python
len(words)
```

```python
from tqdm import tqdm
```

```python
X=[]
for word in tqdm(words): 
    X.append(avg_word2veec(word))
```

```python
X=np.array(X)
```
